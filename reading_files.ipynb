Missing values with na_values parameter



```python
df = pd.read_csv('btc-market-price.csv', header=None, 
                 na_values=['','?','-'])
```

```python
df.head()
```

Column names using names parameter

```python
df = pd.read_csv('btc-market-price.csv', header=None, 
                 na_values=['','?','-'],
                 names=['Timestamp', 'Price'])
```

```python
df.head()
```

```python
df.info()
```

Column types using dtype parameter

```python
df = pd.read_csv('btc-market-price.csv', header=None, 
                 na_values=['','?','-'],
                 names=['Timestamp', 'Price'],
                 dtype={'Price' : 'float'})
```

```python
df.head()
```
